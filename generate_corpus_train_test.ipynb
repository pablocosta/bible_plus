```python
import pandas as pd


df = pd.read_parquet("./data/formal_informal.parquet")
```

```python
df_pt = df[df["language"] == "PT"].sample(43753, random_state=14, replace=True)
df_en = df[df["language"] == "EN"].sample(43753, random_state=14, replace=True)
df_es = df[df["language"] == "ES"].sample(43753, random_state=14, replace=True)
df_it = df[df["language"] == "IT"].sample(43753, random_state=14, replace=True)
df_fr = df[df["language"] == "FR"].sample(43753, random_state=14, replace=True)
```

```python
def write_file(path, data, collumn, lang):
    tok = MosesDetokenizer(lang)
    print(path)
    with open(path, "a+", encoding="utf-8") as outfile:
        outfile.write("\n".join([tok(ele.split(" ")) for ele in data[collumn].to_list()]))

```
